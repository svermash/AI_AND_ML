```python
# algorithm
# 1. initialize input and output
# 2. initialize weight and bias and learning_Rate
# 3. for each input calculate net input(weighted sum)
# 4. apply step function to predict output
# 5. compare predicted_output with real_output
# 6. if its incorrect ,update weight
# 7. repeat until all predicted_output = real_output
# 8. stop
```

```python
#import numpy library for arrays
import numpy as np
```

```python
#input
x1 = np.array([0,0,1,1])
x2 = np.array([0,1,0,1])
#output
op = np.array([0,1,1,1])
```

```python
#initialie weight
w1 = 0.0
w2 = 0.0
w0 = 0.0
x0=1
#learn rate
l_alpha =0.1

made_mistake =True # for the first time we think there will be mistake
while(made_mistake):
  made_mistake =False # then we make it immediately no mistake
  for i in range(len(x1)):
    y_in = x1[i]*w1 + x2[i]*w2 + x0*w0
    if(y_in >=0):
      pred_op =1
      if (pred_op !=op[i]):
        #weight change
        made_mistake =True # we made mistake ,loop again
        print("its inccorect its ",op[i], "not " ,pred_op)

        w1 = w1 + l_alpha*(op[i]-pred_op)*x1[i]
        w2 = w2 + l_alpha*(op[i]-pred_op)*x2[i]
        w0 = w0 + l_alpha*(op[i]-pred_op)*x0
      else:
        print("its correct")
    else:
      pred_op =0
      if(pred_op !=op[i]):
        #weight change
        made_mistake =True # we made mistake, loop again
        print("its inccorect its ",op[i], "not " ,pred_op)

        w1 = w1 + l_alpha*(op[i]-pred_op)*x1[i]
        w2 = w2 + l_alpha*(op[i]-pred_op)*x2[i]
        w0 = w0 + l_alpha*(op[i]-pred_op)*x0
      else:
        print("its correct")
  print("------------")
  print("w1: ",w1)
  print("w2: ",w2)
  print("w0(bias): ",w0)
  print("|--------------------------------|")
```
