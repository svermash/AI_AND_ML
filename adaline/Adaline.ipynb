#Implement Adaline algorithm for AND operation

```python
import numpy as np
```

```python
#input
x1 = np.array([0,0,1,1])
x2 = np.array([0,1,0,1])

#output
op = np.array([0,0,0,1])
```

```python
#initialize weights
w1 = 0.0
w2 = 0.0
w0 = 0.0
x0 = 1.0
alpha =0.01 #learning rate

for k in range(1000):
  for i in range(len(x1)):
    #net input
    y_in = x1[i]*w1 + x2[i]*w2 + x0*w0
    #weight update
    w1 = w1 + alpha*(op[i]-y_in)*x1[i]
    w2 = w2 + alpha*(op[i]-y_in)*x2[i]
    w0 = w0 + alpha*(op[i]-y_in)*x0
print("final weights: ")
print("w1: ",w1)
print("w2: " ,w2)
print("w0: " ,w0)
print('---------------------')

#prediction
for j in range(len(x1)):
  new_y_in = x1[j]*w1 + x2[j]*w2 + x0*w0
  if(new_y_in>=0.5):
    pred_op =1
    print(f"actual {op[j]} | predicted-->{pred_op}")
  else:
    pred_op=0
    print(f"actual {op[j]} | predicted-->{pred_op}")
```
