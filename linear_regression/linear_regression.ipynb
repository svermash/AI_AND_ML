#linear regression

```python
#linear regression --> for prediction
import numpy as np

x= np.array([6,4,1,7,9,2,5]) #study_hours --> independent #input
y=np.array([66,39,7,60,85,30,40]) #marks -->depenedent #output
```

```python
import matplotlib.pyplot as pl
pl.scatter(x,y)
pl.xlabel("study_hours #input")
pl.ylabel("marks #output")
pl.show()
```

#USING STATS (closed -form)

```python
#mean
print(x)
```

```python
sum_x=0
```

```python
#for loop
for i in x:   # for each item in somehting:
                  #do something
  sum_x= sum_x+i
  #print(mean_x)
print(sum_x)
```

```python
print(len(x))
```

```python
mean_x = sum_x/len(x)
print("mean_x:",mean_x)
```

```python
import matplotlib.pyplot as pl
pl.scatter(x,y)
pl.axvline(x=mean_x, color='r', linestyle='--', label='Mean Hours Studied')
pl.xlabel("study_hours")
pl.ylabel("marks")
pl.show()
```

```python
#mean of y
sum_y=0
for j in y:
  sum_y= sum_y+j
print(sum_y)
```

```python
mean_y= sum_y/len(y)
print("mean_y: ",mean_y)
```

```python
import matplotlib.pyplot as pl
pl.scatter(x,y)
pl.axvline(x=mean_x, color='r', linestyle='--', label='Mean Hours Studied')
pl.axhline(y=mean_y, color='g', linestyle='--', label='Mean Score')
pl.xlabel("study_hours")
pl.ylabel("marks")
pl.show()
```

```python
for k in range(len(x)):
  print(x[k]-mean_x)
```

```python
for l in range(len(y)):
  print(y[l]-mean_y)
```

```python
#multiplication of x and y
for i in range(len(x)):
    print ((x[i] - mean_x) * (y[i] - mean_y))
```

```python
# sum of that multiplilcation
mult_x_y = 0
for i in range(len(x)):
    mult_x_y += (x[i] - mean_x) * (y[i] - mean_y)
    print(mult_x_y)
```

```python
print("mult_x_y: ",mult_x_y)
```

```python
sum_xi_x_mean_square=0;
for k in range(len(x)):
  sum_xi_x_mean_square= sum_xi_x_mean_square + (x[k]-mean_x) * (x[k]-mean_x)
print("sum_xi_x_mean_square: ",sum_xi_x_mean_square)
```

```python
slope_b1 =mult_x_y/sum_xi_x_mean_square
print("slope: ", slope_b1)
```

```python
y_intercept_b0 = (mean_y) - (slope_b1*mean_x)
print("y_intercept: ", y_intercept_b0)
```

```python
# y^
for i in range(len(x)):
  print(y_intercept_b0+slope_b1*x[i])
```

```python
#SSE --> sum (yi-y^)^2
#1. (yi-y^)
for m in range(len(x)):
  print(y[m] - (y_intercept_b0+slope_b1*x[m]))
```

```python
#2. (yi-y^)^2
for c in range(len(x)):
  print((y[c] - (y_intercept_b0+slope_b1*x[c]))*(y[c] - (y_intercept_b0+slope_b1*x[c])))
```

```python
#sum (yi-y^)^2
#error SSE
sum_SSE=0
for b in range(len(x)):
  sum_SSE= sum_SSE + ((y[b] - (y_intercept_b0+slope_b1*x[b]))*(y[b] - (y_intercept_b0+slope_b1*x[b])))
print("sum_SSE: ",sum_SSE)
```

```python
#total variation SST
sum_yi_y_mean_square=0;
for k in range(len(x)):
  sum_yi_y_mean_square= sum_yi_y_mean_square + (y[k]-mean_y) * (y[k]-mean_y)
print("sum_yi_y_mean_square: ",sum_yi_y_mean_square)
```

```python
#SSE/SST
SSE_SST = sum_SSE/sum_yi_y_mean_square
print(SSE_SST)
```

```python
#r^2
R_square = 1- SSE_SST
print(R_square)
```

```python
#mse
print("MSE :",sum_SSE/(len(x) -2 )) #2--> coz 2 parameter in data b0 and b1
```

#prediction stats

```python
#prediction
inp= int(input("enter x (study_hours): "))
eqn= y_intercept_b0 + slope_b1 *inp
print("predicted marks: ",eqn)
```

```python
#using logic

import numpy as np
x= np.array([6,4,1,7,9,2,5]) #study_hours --> independent #input
y=np.array([66,39,7,60,85,30,40]) #marks -->depenedent #output
print("x: ",x)
print("y: ",y)
```

```python
import matplotlib.pyplot as pl
pl.scatter(x,y)
pl.xlabel("study_hours #input")
pl.ylabel("marks #output")
pl.show()
```

```python
#initialize b0 & b1 as random
y_intercept_b0 =0
slope_b1=0
```

```python
SUM_MSE= 0
for i in range(len(x)):
  SUM_MSE = SUM_MSE + ((y[i] - (y_intercept_b0+slope_b1 * x[i])) * (y[i] - (y_intercept_b0+slope_b1 * x[i])))
print(SUM_MSE)
```

```python
SUM_MSE= 0
for i in range(len(x)):
  SUM_MSE = SUM_MSE + ((y[i] - (y_intercept_b0+slope_b1 * x[i])) **2)
print(SUM_MSE)
```

```python
MSE= SUM_MSE/len(x)
print("MSE: ",MSE)
```

```python
# update rule
y_intercept_b0 =0
slope_b1=0
for j in range(1000): #1000 epoch

  sum_yi_pred= 0
  sum_xi_yi_pred= 0
  alpha =0.001
  for i in range(len(x)):
    sum_yi_pred= sum_yi_pred + y[i]- (y_intercept_b0+slope_b1 * x[i])
    sum_xi_yi_pred =sum_xi_yi_pred + x[i] * (y[i]- (y_intercept_b0+slope_b1 * x[i]))

  mult_2_sum_yi_pred = (-2/len(x)) * sum_yi_pred
  mult_2_sum_xi_yi_pred = (-2/len(x)) * sum_xi_yi_pred

  y_intercept_b0 = y_intercept_b0 - alpha*mult_2_sum_yi_pred
  slope_b1 = slope_b1 - alpha*mult_2_sum_xi_yi_pred

print(y_intercept_b0)
print(slope_b1)
```

```python
import matplotlib.pyplot as pl
pl.scatter(x,y)
pl.xlabel("study_hours #input")
pl.ylabel("marks #output")
pl.show()
```

```python
import matplotlib.pyplot as pl

pl.scatter(x, y)

# predicted line
y_pred = []
for i in range(len(x)):
    y_pred.append(y_intercept_b0 + slope_b1 * x[i])

pl.plot(x, y_pred,color='red')

pl.xlabel("study_hours #input")
pl.plot(0, y_intercept_b0, marker='o' ,color='green')
pl.ylabel("marks #output")
pl.show()
```
